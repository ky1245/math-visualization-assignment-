## Task 1 — Mathematical Function Visualization


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 200)
y1 = x
y2 = x**2
y3 = np.sin(x)
y4 = np.exp(-0.1 * x) * np.cos(x)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='y = x', linestyle='-', color='blue')
plt.plot(x, y2, label='y = x²', linestyle='--', color='red')
plt.plot(x, y3, label='y = sin(x)', linestyle=':', color='green')
plt.plot(x, y4, label='y = e^(-0.1x) * cos(x)', linestyle='-.', color='purple')

plt.title('Plot of Four Functions')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend()
plt.grid(True)
plt.savefig('function_plot.png')

plt.show()
```

## Task 2 — Your Own Equation

```python
y_another_custom = x * np.cos(x) + (x**2 / 2) - 5

plt.figure(figsize=(10, 6))

plt.plot(x, y_another_custom, label='y = x * cos(x) + x²/2 - 5', color='darkgreen')

plt.title('Custom Equation Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend()
plt.grid(True)
plt.savefig('own_equation.png')
print("This plot visualizes the custom equation: y = x * cos(x) + x**2 / 2 - 5")

plt.show()
```

## Task 3 — Student Score Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

students = ["S1", "S2", "S3", "S4", "S5", "S6", "S7", "S8", "S9", "S10"]
midterm = np.array([85, 72, 90, 66, 78, 92, 60, 74, 88, 95])
final = np.array([80, 70, 94, 68, 75, 90, 65, 72, 84, 96])

total = 0.4 * midterm + 0.6 * final

df_scores = pd.DataFrame({
    'Student': students,
    'Midterm': midterm,
    'Final': final,
    'Total': total
})
display(df_scores.head())
```

### A. Scatter Plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_scores['Midterm'], df_scores['Final'], color='teal', s=100, alpha=0.8)
plt.title('Midterm vs. Final Score')
plt.xlabel('Midterm Score')
plt.ylabel('Final Score')
plt.grid(True)
plt.savefig('score_scatter.png')
plt.show()
```

### B. Histogram

```python
plt.figure(figsize=(8, 6))
plt.hist(df_scores['Total'], bins=5, edgecolor='black', alpha=0.7, color='lightcoral')
plt.title('Distribution of Total Scores')
plt.xlabel('Total Score')
plt.ylabel('Number of Students')
plt.grid(axis='y', alpha=0.75)
plt.savefig('score_histogram.png')
plt.show()
```

### C. Bar Chart

```python
plt.figure(figsize=(10, 7))
plt.bar(df_scores['Student'], df_scores['Total'], color='cornflowerblue', alpha=0.8)
plt.title('Total Score per Student')
plt.xlabel('Student')
plt.ylabel('Total Score')
plt.ylim(0, 100)
plt.grid(axis='y', alpha=0.75)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('score_bar_chart.png')
plt.show()
```

## Task 4— Best-Fit Line / Simple Prediction

```python
slope, intercept = np.polyfit(df_scores['Midterm'], df_scores['Final'], 1)

x_fit = np.linspace(df_scores['Midterm'].min(), df_scores['Midterm'].max(), 100)
y_fit = slope * x_fit + intercept

plt.figure(figsize=(9, 7))
plt.scatter(df_scores['Midterm'], df_scores['Final'], color='blue', label='Original Data', s=80, alpha=0.7)
plt.plot(x_fit, y_fit, color='red', label=f'Best-Fit Line (y = {slope:.2f}x + {intercept:.2f})')

plt.title('Midterm vs. Final Score with Best-Fit Line')
plt.xlabel('Midterm Score')
plt.ylabel('Final Score')
plt.legend()
plt.grid(True)
plt.savefig('score_prediction.png')
plt.show()
```

### Print Prediction Examples

```python
print("Predicted Final Scores:")
for midterm_score in [50, 75, 100]:
    predicted_final = slope * midterm_score + intercept
    print(f"  Midterm Score {midterm_score}: Predicted Final Score = {predicted_final:.2f}")
```
